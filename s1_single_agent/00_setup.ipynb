# Azure AI Agent with MCP Tools
This notebook provides setup and usage instructions for the Azure AI Agent repository, including environment setup, configuration, and Azure login.

## Step 1. Environment Setup

### Option 1: Typical Setup
1. Create a virtual environment:
   ```bash
   python -m venv .venv
   ```
2. Activate the virtual environment:
   * **Windows PowerShell**
        ```bash
        .venv\Scripts\Activate
        ```
   * **CMD**
        ```bash
        .venv\Scripts\activate.bat
        ```
   * **macOS/Linux**
        ```bash
        source .venv/bin/activate
        ```
3. Install required packages:
   ```bash
   pip install -r requirements.txt
   ```

### Option 2: Using `uv sync`
1. Quickly set up and install the virtual environment:
   ```bash
   uv sync
   ```
2. Activate the virtual environment:
   * **Windows PowerShell**
        ```bash
        .venv\Scripts\Activate
        ```
   * **CMD**
        ```bash
        .venv\Scripts\activate.bat
        ```
   * **macOS/Linux**
        ```bash
        source .venv/bin/activate
        ```

## Step 3. Azure Login (az login)
Before running the scripts, please log in to your Azure account using the Azure CLI. Use the following command **in your terminal**:

1. Install the Azure CLI by following the instructions [How to install the Azure CLI](https://learn.microsoft.com/en-us/cli/azure/install-azure-cli?view=azure-cli-latest)

2. Log in to your Azure account.
    - **Windows (PowerShell or CMD):**
        ```bash
        az login
        ```

    - **macOS / Linux (Terminal):**
        ```bash
        az login
        ```

This will open a browser window for you to authenticate with your Azure account. After successful login, you can proceed with the next steps.
